```python
import os
from bson import ObjectId
from IPython.display import Image
from pprint import pprint
stream = os.popen('ls -1 temp/*.jpg temp/*.JPG')
names = [f for f in stream.read().split("\n") if f != '']
pprint(names)
```

```python
for n in names: 
    print(str(ObjectId())+'.jpg')
    os.popen('mv {} {}'.format(n, '../data/'+str(ObjectId())+'.jpg')) 
```
